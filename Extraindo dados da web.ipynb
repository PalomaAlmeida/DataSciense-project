```python
import csv
from bs4 import BeautifulSoup
import requests
```

```python
def scrape(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #Pegando as divs com o conteudo dos livros
    books = soup.find_all("div", class_="prateleiraProduto__informacao")

    #Cria o arquivo
    with open('books_data_test.csv', 'a', newline='', encoding='UTF-16') as f:
        f.write('Título | Autor | Categoria | Preço \n')
        print('Adicionando livros...')
        for book in books:
            #Buscando os dados 
            title = book.find("h2", class_="prateleiraProduto__informacao__nome").get_text().strip()

            colaborador = book.find("div", class_="prateleiraProduto__autor__nome").get_text().strip()
            author = colaborador[12:]
            
            category = book.find("div", class_="prateleiraProduto__informacao__categoria").get_text().strip()
            
            try:
                price = book.find("span", class_="prateleiraProduto__informacao__preco--valor").get_text().strip()
                num_price = price[3:]
            except:
                num_price = '0'

            # Grava as linhas
            linha = title + ' | ' + author.strip() + ' | ' + category + ' | ' + num_price + '\n'
            print(linha)
            f.write(linha)  
        print('Livros adicionados!')
```

```python
def main():
    print("Salvando a saída das informações extraídas")
    with open("url.csv") as file:
        csv_reader = csv.DictReader(file)
        for csv_row in csv_reader:
            for i in range(1,5):
                url = csv_row['url']+f'{i}'
                print(url)
                scrape(url)
```

```python
main()
```
